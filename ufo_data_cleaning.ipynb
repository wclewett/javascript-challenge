```python
# import dependencies
import pandas as pd
import json
```

```python
# read in dirty ufo data from data world
path = 'extra-data/nuforc_reports.json'
ufo_data_new = pd.read_json(path, lines=True)
ufo_data_new
```

```python
# column info
ufo_data_new.info()
```

```python
# drop na's
ufo_data_new = ufo_data_new.dropna()
ufo_data_new = ufo_data_new.reset_index(drop=True)
ufo_data_new.info()
```

```python
ufo_data_json = []

# Remove records that are outside the us and store the remaining in json format
for i in range(len(ufo_data_new['date_time'])):
    if ufo_data_new['city'][i].find('(') < 0:
    # create dictionary level for each record
        ufo_data = {}
        ufo_data['datetime'] = ''
        ufo_data['city'] = ''
        ufo_data['state'] = ''
        ufo_data['country'] = ''
        ufo_data['shape'] = ''
        ufo_data['durationMinutes'] = ''
        ufo_data['comments'] = ''
        
    # add data to dictionary levels
        ufo_data['datetime'] = ufo_data_new['date_time'][i][0:8]
        ufo_data['city'] = ufo_data_new['city'][i].lower()
        ufo_data['state'] = ufo_data_new['state'][i].lower()
        ufo_data['country'] = 'us'
        ufo_data['shape'] = ufo_data_new['shape'][i].lower()
        ufo_data['durationMinutes'] = ufo_data_new['duration'][i].lower()
        ufo_data['comments'] = ufo_data_new['summary'][i]
    
    # add dictionary to json list
        ufo_data_json.append(ufo_data)
        
```

```python
ufo_data_json
```

```python
# write the dictionary as a text file to be copied to a data.js file
with open('static/js/data.js', 'w') as out_file:
    out_file.write('var ufo_data = %s;' % ufo_data_json)
```
